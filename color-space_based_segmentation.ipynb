```python
import cv2
import numpy as np

import matplotlib
#matplotlib.use('nbagg')
import matplotlib.pylab as plt

import ipywidgets as widgets
```

```python
hand = cv2.imread("./hand-01.jpg")
# plt.imshow(cv2.cvtColor(hand, cv2.COLOR_BGR2RGB))
# plt.axis('off')
# plt.show()
```

## Segmentation using HSV

```python
hand_hsv = cv2.cvtColor(hand, cv2.COLOR_BGR2HSV)
lower = np.array((0, 10, 60))
upper = np.array((20, 150, 255))

hand_skin = cv2.inRange(hand_hsv,lower, upper)
plot_obj = plt.imshow(hand_skin)
plt.axis('off')

# Adds interactive elements based on ipywidgets
def hue_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[0] = v_lower
        upper[0] = v_upper
   
        hand_skin = cv2.inRange(hand_hsv,lower, upper)
        
        plot_obj.set_data(hand_skin)
        plot_obj.axes.figure.canvas.draw()
        
        
def sat_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[1] = v_lower
        upper[1] = v_upper
   
        hand_skin = cv2.inRange(hand_hsv,lower, upper)
        
        plot_obj.set_data(hand_skin)
        plot_obj.axes.figure.canvas.draw()
        

def val_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[2] = v_lower
        upper[2] = v_upper
   
        hand_skin = cv2.inRange(hand_hsv,lower, upper)
        
        plot_obj.set_data(hand_skin)
        plot_obj.axes.figure.canvas.draw()

slider_hue = widgets.IntRangeSlider(value=(lower[0], upper[0]), min=0, max=255, description='Hue')
slider_hue.observe(hue_handler)
display(slider_hue)

slider_sat = widgets.IntRangeSlider(value=(lower[1], upper[1]), min=0, max=255, description='Sat')
slider_sat.observe(sat_handler)
display(slider_sat)

slider_val = widgets.IntRangeSlider(value=(lower[2], upper[2]), min=0, max=255, description='Val')
slider_val.observe(val_handler)
display(slider_val)
```

## Segmentation using YCrCb

```python
hand_ycrcb = cv2.cvtColor(hand, cv2.COLOR_BGR2YCrCb)
lower = np.array((0, 133, 77))
upper = np.array((255, 173, 177))

hand_skin2 = cv2.inRange(hand_hsv,lower, upper)
plot_obj2 = plt.imshow(hand_skin)
plt.axis('off')

# Adds interactive elements based on ipywidgets
def y_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[0] = v_lower
        upper[0] = v_upper
   
        hand_skin2 = cv2.inRange(hand_ycrcb,lower, upper)
        
        plot_obj2.set_data(hand_skin2)
        plot_obj2.axes.figure.canvas.draw()
        
        
def cr_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[1] = v_lower
        upper[1] = v_upper
   
        hand_skin2 = cv2.inRange(hand_ycrcb,lower, upper)
        
        plot_obj2.set_data(hand_skin2)
        plot_obj2.axes.figure.canvas.draw()
        

def cb_handler(change):
    new_values_dict = change['new']
    if isinstance(new_values_dict, dict) and new_values_dict.get('value'):
        v_lower, v_upper = new_values_dict.get('value')
        
        lower[2] = v_lower
        upper[2] = v_upper
   
        hand_skin2 = cv2.inRange(hand_ycrcb,lower, upper)
        
        plot_obj2.set_data(hand_skin2)
        plot_obj2.axes.figure.canvas.draw()

slider_y = widgets.IntRangeSlider(value=(lower[0], upper[0]), min=0, max=255, description='Y')
slider_y.observe(y_handler)
display(slider_y)

slider_cr = widgets.IntRangeSlider(value=(lower[1], upper[1]), min=0, max=255, description='Cr')
slider_cr.observe(cr_handler)
display(slider_cr)

slider_cb = widgets.IntRangeSlider(value=(lower[2], upper[2]), min=0, max=255, description='Cb')
slider_cb.observe(cb_handler)
display(slider_cb)
```
